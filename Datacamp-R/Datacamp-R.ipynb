```r
# Matrix Arithmetic

the_fellowship = c(316, 556)
two_towers = c(343, 584)
return_king = c(378, 742)

lotr_matrix = rbind(the_fellowship, two_towers, return_king)
colnames(lotr_matrix) = c("US", "non-US")
rownames(lotr_matrix) = c("Fellowship", "Two Towers", "Return King")

print(lotr_matrix)
```

```r
lotr_matrix / 1.05
```

```r
# Categorical Variables 

# Create factor

blood = c("B", "AB", "O", "A", "O", "O", "A", "B")

print(blood)

blood_factor = factor(blood)

print(blood_factor)
```

```r
str(blood_factor)
```

```r
# Orders levels differently

blood_factor2 = factor(blood, levels = c("O", "A", "B", "AB"))

print(blood_factor2)

str(blood_factor2)
```

```r
# Rename factor levels

blood = c("B", "AB", "O", "A", "O", "O", "A", "B")

blood_factor = factor(blood)

levels(blood_factor) = c("BT_A", "BT_AB", "BT_B", "BT_O")

print(blood_factor)
```

```r
factor(blood, labels = c("BT_A", "BT_AB", "BT_B", "BT_O"))
```

```r
factor(blood,
      levels = c("O", "A", "B", "AB"), 
      labels = c("BT_O", "BT_A", "BT_B", "BT_AB"))
```

```r
# Nominal versus Ordinal

blood_factor[1] < blood_factor[2]
```

```r
tshirt = c("M", "L", "S", "S", "L", "M", "L", "M")

tshirt_factor = factor(tshirt, ordered = TRUE, levels = c("S", "M", "L"))

print(tshirt_factor)
```

```r
tshirt_factor[1] < tshirt_factor[2]
```

```r
# Lists in R  (can contain all types of objects)

song = list("Rsome", 190, 5)

print(song)

is.list(song)

names(song) = c("title", "duration", "track")

print(names(song))
```

```r
# Alternative

song = list(title = "Rsome times", duration = 190, track = 5)

str(song)
```

```r
# List in List

similar_song = list(title = "R you on time?",
                   duration = 230)

song = list(title = "Rsome times", duration = 190, track = 5,
           similar = similar_song)

str(song)
```

```r
song[1]

song[[1]]          # we need double bracket to obtain elements in the list

song[c(1,3)]
```

```r
# Subsetting songs list

song[[c(1, 1)]]

song[[2]][[1]]          # these two commands are equivalent

print(song[[4]][[1]])
print(song[[4]][[2]])
```

```r
# subsetting by names

song[["duration"]]

song["duration"]

song[c("duration", "similar")]
```

```r
# Subset by logical : only single bracket works

song[c(F,T,T,F)]
```

```r
# $ and extending

song$duration
```

```r
friends = c("Kurt", "Florence", "Patti", "Dave")

song$sent = friends

print(song)
```

```r
song[["sent"]] = friends

song$similar$reason = "too long"

song
```
