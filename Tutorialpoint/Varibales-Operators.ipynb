# R-Operators

```r
# Add two vectors

v <- c(2,5.5, 6)
t <- c(8,3,4)

print(v+t)
```

```r
# Subtract two vectors

v <- c(2, 5.5, 6)
t <- c(8, 3, 4)

print(v-t)
```

```r
# Divide first vector by the second

v <- c(2,5.5,6)

t <- c(8,3,4)

print(v/t)
```

```r
# Remainder

print(v %% t)
```

```r
# Quotient

print(v %/% t)
```

```r
# Exponent

print(v^t)
```

## Relational Operators

```r
print(v>t)

print(v<t)

print(v==t)

print(v<=t)

print(v>=t)

print(v != t)
```

## Logical Operators

```r
# AND Operator

v <- c(3, 1, TRUE, 2+3i)
t <- c(4, 1, FALSE, 2+3i)

print(v&t)
```

```r
# OR Operator

v <- c(3, 0, TRUE, 2+2i)
t <- c(4, 0, FALSE, 2+3i)

print(v|t)
```

```r
# NOT Operator

v <- c(3,0,TRUE,2+2i)

print(!v)
```

```r
# Logical AND Operator

v <- c(3, 1, TRUE, 2+3i)
t <- c(4, 1, FALSE, 2+3i)

print(v&&t)     # Takes first element only
```

```r
# Logical OR Operator

v <- c(3, 0, TRUE, 2+2i)
t <- c(4, 0, FALSE, 2+3i)

print(v||t)    # Takes first elements only of both the vectors
```

## Assignment Operator

```r
# Left Assignment

v1 <- c(3,1,TRUE,2+3i)
v2 <<- c(3,1,TRUE,2+3i)
v3 = c(3,1,TRUE,2+3i)

print(v1)
print(v2)
print(v3)
```

```r
# Right Assignment

c(3,1,TRUE,2+3i) -> v1
c(3,1,TRUE,2+3i) ->> v2

print(v1)
print(v2)
```

## Miscellaneous Operators

```r
# Colon Operator : 

v <- 2:8
print(v)       # Creates a series of numbers in sequence for a vector
```

```r
#  %in% Operator

v1 <- 8
v2 <- 12
t <- 1:10

print(v1 %in% t)   # Identify if element belongs to a vector

print(v2 %in% t)
```

```r
#  %*% Operator

M = matrix( c(2,6,5,1,10,4), nrow = 2, ncol = 3, byrow = TRUE)

t = M %*% t(M)       # Used to multiply a matrix with its transpose

print(t)
```
