# Decision Making

```r
# if Statement

x <- 30L

if (is.integer(x)) {
    print("X is an integer")
}
```

```r
# if else statement

x <- c("what", "is", "truth")

if ("Truth" %in% x) {
    print("Truth is found")
} else {
    print("Truth is not found")
}
```

```r
# if else - if else statement

if ("Truth" %in% x) {
    print("Truth is found the first time")
} else if ("truth" %in% x) {
    print("truth is found the second time")
} else {
    print("No truth found")
}
```

```r
# Switch statement

x <- switch(
    2,
    "first",
    "second",
    "third",
    "fourth"
)

print(x)
```

## R-Loops

```r
# Repeat Loop

v <- c("Hello", "loop")
cnt <- 2

repeat {
    print(v)
    cnt <- cnt + 1
    
    if(cnt > 5) {
        break
    }
}
```

```r
# While loop 

v <- c("Hello", "while loop")

cnt <- 2

while (cnt < 7) {
    print(v)
    cnt = cnt + 1
}
```

```r
# for loop

v <- LETTERS[1:4]
for (i in v) {
    print(i)
}
```

## Loop control statements

```r
# break statement

v <- c("Hello R")
cnt <- 1

repeat {
    print(v)
    cnt <- cnt + 1
    
    if (cnt > 3) {
        break
    }
}
```

```r
# Next statement

v <- LETTERS[1:6]

for (i in v) {
    
    if (i == 'D') {
        next
    }
    
    print(i)
}
```

## Functions

```r
# Built-in Functions

print(seq(32,44))

print(mean(25:82))

print(sum(41:68))
```

```r
# User defined function

new.function <- function(a){
    for (i in 1:a){
        b <- i^2
        print(b)
    }
}


new.function(6)            # Calling a function
```

```r
# funcition call without argument

new.function <- function(){
    for (i in 1:5){
        b <- i^2
        print(b)
    }
}

new.function()
```

```r
# Calling a function with arguments.

new.function <- function(a,b,c) {
    result <- a*b+c
    print(result)
}

# call the funciton by positoin of arguments

new.function(5,3,9)
```
