```r
# Density based clustering
# Iris data

data('iris')
```

```r
str(iris)
```

```r
new = iris[,-5]    # Remove 5th column
```

```r
# Installing packages

library(fpc)
library(dbscan)
library(factoextra)
```

```r
# Obtaining optimal eps value

kNNdistplot(new, k = 3)

abline(h = 0.45, lty = 2)
```

```r
# Density-based clustering with fpc and dbscan

set.seed(123)

f = fpc::dbscan(new, eps = 0.45, MinPts = 4)

print(f)
```

**o is the noise. 17 points do not belong to any cluster and is noise. There are 3 cluster 1, 2, 3 in the table above.**

```r
d = dbscan(new, 0.45, 4)

print(d)
```

```r
# Cluster visualization

fviz_cluster(f, new, geom = "point")
```

```r
fviz_cluster(f, new, geom = "point")
```

**We get similar output for same parameter for fpc and dbscan. We can chage MinPts with different values and see the different cluster and noise in the cluster distribution**
